```python
import pandas as pd

df=pd.read_csv("C:/Users/SAMA/Desktop/E-Commerce_Customer_Analysis/customer_shopping_behavior.csv")
```

```python
# Impute missing Review Rating values with the median rating of their Category
df['Review Rating'] = df['Review Rating'].fillna(
    df.groupby('Category')['Review Rating'].transform('median')
)
```

```python
df.isnull().sum()
```

```python
df.columns=df.columns.str.lower()
df.columns=df.columns.str.replace(' ','_')
```

```python
df
```

```python
df=df.rename(columns={'purchase_amount_(usd)':'purchase_amount'})
```

```python
df.columns
```

```python
labels=['Young','Adult','Middle-Age','Old']
df['age_group']=pd.qcut(df['age'],q=4, labels=labels)
```

```python
df
```

```python
df[['age','age_group']].head(10)
```

```python
# create new column purchase_frequency_days

frequency_mapping = {
    'Fortnightly': 14,
    'Weekly': 7,
    'Monthly': 30,
    'Quarterly': 90,
    'Bi-Weekly': 14,
    'Annually': 365,
    'Every 3 Months': 90
}

df['purchase_frequency_days'] = df['frequency_of_purchases'].map(frequency_mapping)
```

```python
df[['purchase_frequency_days', 'frequency_of_purchases']].tail(10)
```

```python
(df['discount_applied'] == df['promo_code_used']).all()
```

```python
# axis=1 = drop column
# axis=0 = drop row
df = df.drop('promo_code_used', axis=1)
```

```python
df.columns
```

```python
# To not include the DataFrame’s index as a separate column
df.to_csv('customer_shopping_behaviour_cleaned_data.csv', index=False)
```
